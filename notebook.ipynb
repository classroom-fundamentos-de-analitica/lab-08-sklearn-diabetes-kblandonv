```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo CSV
df_test = pd.read_csv('test_dataset.csv')
df_train = pd.read_csv('train_dataset.csv')
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Cargar los datasets
train_df = pd.read_csv('train_dataset.csv')
test_df = pd.read_csv('test_dataset.csv')

# Separar características (X) y variable objetivo (y) en ambos datasets
X_train = train_df.drop(columns=['target'])  # Reemplaza 'target' con el nombre de la columna objetivo
y_train = train_df['target']
X_test = test_df.drop(columns=['target'])    # Reemplaza 'target' con el nombre de la columna objetivo
y_test = test_df['target']

# Ajustar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de entrenamiento
y_train_pred = model.predict(X_train)

# Hacer predicciones en el conjunto de prueba
y_test_pred = model.predict(X_test)


# Calcular las métricas de rendimiento
mse_test = mean_squared_error(y_test, y_test_pred)
mae_test = mean_absolute_error(y_test, y_test_pred)
r2_test = r2_score(y_test, y_test_pred)

mse_train = mean_squared_error(y_train, y_train_pred)
mae_train = mean_absolute_error(y_train, y_train_pred)
r2_train = r2_score(y_train, y_train_pred)

# Imprimir las métricas para el conjunto de entrenamiento
print('Métricas para el conjunto de entrenamiento:')
print(f'Mean Squared Error (MSE): {mse_train:.4f}')
print(f'Mean Absolute Error (MAE): {mae_train:.4f}')
print(f'R-squared (R2): {r2_train:.4f}')

# Imprimir las métricas para el conjunto de prueba
print('\nMétricas para el conjunto de prueba:')
print(f'Mean Squared Error (MSE): {mse_test:.4f}')
print(f'Mean Absolute Error (MAE): {mae_test:.4f}')
print(f'R-squared (R2): {r2_test:.4f}')
```

```python
import pickle

# Guardar el modelo entrenado usando pickle
with open('model.pkl', 'wb') as file:
    pickle.dump(model, file)
```
